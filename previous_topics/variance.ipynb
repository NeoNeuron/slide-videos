# 6-$\sigma$-3-$\sigma$ Demo

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
from scipy.stats import t,chi2,norm

def normal_dis(x, mu, sigma=0.2*0.2/25):
    '''
    x: 输入待计算的值
    mu: (mu, 读作：谬) 为数学期望, 
    sigma: (sigma, 读作：西格玛）为标准差
    '''
    return norm.pdf(x,mu,sigma)

# pyplot methods reference
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html?highlight=plot#module-matplotlib.pyplot 

# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

```

## 不共享x轴

```python
#6sigma

# mu 和 sigma
mu, sigma = 50,1
xs = [x for x in np.linspace(44,56,2000)]
ys = [normal_dis(x, mu, sigma) for x in xs]

plt.figure(figsize=(15, 5),dpi=200)
plt.plot(xs, ys, color='b')

plt.ylim(0,0.5)

num_sigma = 6#需要画sigma的个数

for i in range(1, num_sigma+1):  
    # 计算中间三个区间连线的两端坐标
    x1 = mu - sigma * i *0.5
    x2 = mu + sigma * i *0.5
    y1 = normal_dis(x1, mu, sigma)
    y2 = normal_dis(x2, mu, sigma)
    
    plt.plot([x1,x1],[0,y1],color='b',linestyle='--')
    plt.plot([x2,x2],[0,y2],color='b',linestyle='--')
    
plt.plot([mu,mu],[0,normal_dis(mu,mu,sigma)],color='b',linestyle='--')

plt.yticks([])  # 去掉y轴

#构建x轴
x_ticks = [(mu - sigma * (num_sigma+1-i) *0.5) for i in range(1, num_sigma+1)]+[(mu + sigma * (i) *0.5) for i in range(1, num_sigma+1)]
x_values = [('-' + str(num_sigma+1-i) + r'$\sigma$') for i in range(1, num_sigma+1)] + [('+' + str(i) + r'$\sigma$') for i in range(1, num_sigma+1)]
plt.xticks(x_ticks+[mu],x_values+[r'$\mu$'],fontsize=12,color='b')#绘制sigma和mu的坐标


#3sigma

# mu 和 sigma
mu, sigma = 50,2
xs = [x for x in np.linspace(44,56,2000)]
ys = [normal_dis(x, mu, sigma) for x in xs]

plt.figure(figsize=(15, 5),dpi=200)
plt.plot(xs, ys, color='b')

plt.ylim(0,0.5)

num_sigma = 3#需要画sigma的个数

for i in range(1, num_sigma+1):  
    # 计算中间三个区间连线的两端坐标
    x1 = mu - sigma * i *0.5
    x2 = mu + sigma * i *0.5
    y1 = normal_dis(x1, mu, sigma)
    y2 = normal_dis(x2, mu, sigma)
    
    plt.plot([x1,x1],[0,y1],color='b',linestyle='--')
    plt.plot([x2,x2],[0,y2],color='b',linestyle='--')

plt.plot([mu,mu],[0,normal_dis(mu,mu,sigma)],color='b',linestyle='--')

plt.yticks([])  # 去掉y轴

#构建x轴
x_ticks = [(mu - sigma * (num_sigma+1-i) *0.5) for i in range(1, num_sigma+1)]+[(mu + sigma * (i) *0.5) for i in range(1, num_sigma+1)]
x_values = [('-' + str(num_sigma+1-i) + r'$\sigma$') for i in range(1, num_sigma+1)] + [('+' + str(i) + r'$\sigma$') for i in range(1, num_sigma+1)]
plt.xticks(x_ticks+[mu],x_values+[r'$\mu$'],fontsize=12,color='b')#绘制sigma和mu的坐标
```

# 投资组合决策边界示意图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sco
from scipy.optimize import minimize

e = 1e-10 # 非常接近0的值
fun = lambda x : (55 * x[0] * x[0] + 28 * x[1] * x[1]) # 约束函数
cons = ({'type': 'eq', 'fun': lambda x: 60 *x[0] + 48 * x[1] +x[2] - 1e5}, 
        {'type': 'ineq', 'fun': lambda x:6 *x[0] + 4 * x[1] + 0.036 * x[2] - e - 7e3}# x>=e，即 x > 0
       )
x0 = np.array((500, 600, 41200)) # 设置初始值
res = minimize(fun, x0, method='SLSQP', constraints=cons)
print('最小值：',res.fun)
print('最优解：',res.x)
print('迭代终止是否成功：', res.success)
print('迭代终止原因：', res.message)
```

```python
returns_annual = np.array([6,4,0.036])
cov_annual = np.array([[55,0,0],[0,28,0],[0,0,0]])
total_money = 1e5

best_ER = np.dot(res.x,returns_annual)
best_DR = np.dot(res.x.T, np.dot(cov_annual, res.x))/1e7#标准化
```

```python
np.random.seed(5)

number_assets = 3#资产数量
weights = np.random.random(number_assets)
weights /= np.sum(weights)

portfolio_returns = []
portfolio_volatilities = []
sharpe_ratio = []
for single_portfolio in range(1000000):#模拟1e6次
    weights = np.random.random(number_assets)
    weights /= np.sum(weights) #归一化
    spend_money_on_each_product = total_money * weights
    
    s1 = spend_money_on_each_product[0] / 60.0
    s2 = spend_money_on_each_product[1] / 48.0
    s3 = spend_money_on_each_product[2]
    
    product_ratio = np.array([s1,s2,s3])#s1,s2,s3
    returns = np.dot(product_ratio, returns_annual)#该资产组合期望收益
    volatility = np.dot(product_ratio.T, np.dot(cov_annual, product_ratio))
    
    portfolio_returns.append(returns)
    portfolio_volatilities.append(volatility)
    sharpe = returns / volatility
    sharpe_ratio.append(sharpe)

portfolio_returns = np.array(portfolio_returns)
portfolio_volatilities = np.array(portfolio_volatilities)
```

```python
portfolio_volatilities /= 1e7#画图前先做标准化
```

```python
fig=plt.figure(dpi=200,figsize=(4,4))
ax1 = plt.subplot()

ax1.scatter(portfolio_volatilities, portfolio_returns, marker='.',c='royalblue') # c=sharpe_ratio,edgecolors='black',cmap='RdYlGn',
ax1.set_xlabel(r'$DR$' + r'$(\times 10^7)$')
ax1.set_ylabel(r'$ER$')

#xy轴刻度
x_ticks = range(0,17,2)
x_values = [str(i) for i in x_ticks]

ax1.set_xticks(x_ticks)#绘制sigma和mu的坐标
ax1.set_xticklabels(x_values,fontsize=8)


y_ticks = range(1000,11001,2000)
y_values = [str(i) for i in y_ticks]

ax1.set_yticks(y_ticks)#绘制sigma和mu的坐标
ax1.set_yticklabels(y_values,fontsize=8)


#xy轴范围
ax1.set_xlim(0,16)
ax1.set_ylim(0,12000)

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
```
